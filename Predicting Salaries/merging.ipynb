```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sal = pd.read_csv('salaries.csv')
#andy = pd.read_csv('zip_code_data.csv')
jon = pd.read_csv('US_Zip_Income_Fixed.csv')
#ian1 = pd.read_csv('Zip_Zhvi_Summary_AllHomes.csv')
#ian2 = pd.read_csv('Zip_MedianValuePerSqft_AllHomes.csv')
```

```python
print(sal.shape)
#print(andy.shape)
print(jon.shape)
#print(ian1.shape)
#print(ian2.shape)
```

```python
print(jon.head())
print(jon.dtypes)
```

```python
# Fix Zip variable in jon
jon['ZipCopy'] = jon['Zip']
jon['Zip'] = jon['Zip'].apply(str)
jon['Zip'] = jon['Zip'].str[0:5]
jon['inJon'] = True
print(jon.head(5))
```

```python
# Fix Zip variable in sal
sal['Zip'] = sal['Zip'].str[0:5]
sal['inSal'] = True
sal.head(500)
```

```python
# merge
merged = pd.merge(sal, jon, how='left', on='Zip')
```

```python
print(merged.shape)
print(merged.head(5))
```

```python
merged['inBoth'] = (merged['inJon']==True)
print('Number of Unique Zip Codes in Salary Data: ')
print(merged.Zip.nunique())
print('Table of Merge Matches (=True if in jon AND sal)')
print(merged.groupby('inBoth').Zip.nunique())

print(' ')
print('US ONLY: Number of Unique Zip Codes in Salary Data: ')
print(merged[merged['country'] == 'United States'].Zip.nunique())
print('US ONLY: Table of Merge Matches (=True if in jon AND sal)')
print(merged[merged['country'] == 'United States'].groupby('inBoth').Zip.nunique())
```

```python
# save CSV
merged.to_csv('merged.csv')
```
